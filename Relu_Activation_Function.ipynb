```python
#importing libraries
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#creating Sequential random datapoints(over x-axis)
random_number= np.linspace(-10,10,100)
print(random_number)
```

```python
#creating sequential random datapoints(over y-axis)
def relu_activation(datapoints):
    #creating empty list to contain filtered datapoints
    relu_value = []
    for i in datapoints:
        initial_value = 0
        if i > 0:
            initial_value = i
        relu_value.append(initial_value)
            
    return relu_value

relu_line = relu_activation(random_number)
print(relu_line)
```

```python
#Visualizing the Relu Activation Function
plt.title("Relu Activation Function")
plt.plot(random_number,relu_line)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```
