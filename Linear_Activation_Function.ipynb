```python
#importing libraries
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#creating Sequential random datapoints(over x-axis)
random_number= np.linspace(-10,10,100)
print(random_number)
```

```python
#creating sequential random datapoints(over y-axis)
def linear_activation(datapoints):
    #creating empty list to contain filtered datapoints
    a = []
    for i in datapoints:
        a.append(i)
    return a

linear_line = linear_activation(random_number)
print(linear_line)
```

```python
#Visualizing the Linear Activation Function
plt.title("Linear Activation Function")
plt.plot(random_number,linear_line)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
