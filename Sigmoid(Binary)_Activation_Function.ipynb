```python
#importing libraries
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#creating Sequential random datapoints(over x-axis)
random_number= np.linspace(-10,10,100)
print(random_number)
```

```python
#creating sequential random datapoints(over y-axis)
def sigmoid_activation(datapoints):
    #creating empty list to contain filtered datapoints
    sigmoid_value = []
    for i in datapoints:
        sigmoid = 1/(1+math.exp(-i))
        sigmoid_value.append(sigmoid)
    return sigmoid_value

sigmoid_line = sigmoid_activation(random_number)
print(sigmoid_line)
```

```python
#Visualizing the Sigmoid(Binary) Activation Function
plt.title("Sigmoid(Binary) Activation Function")
plt.plot(random_number,sigmoid_line)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
