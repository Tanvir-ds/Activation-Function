```python
#importing libraries
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#creating Sequential random datapoints(over x-axis)
random_number= np.linspace(-10,10,100)
print(random_number)
```

```python
#creating sequential random datapoints(over y-axis)
def tanh_activation(datapoints):
    #creating empty list to contain filtered datapoints
    tanh_value = []
    for i in datapoints:
        tanh = math.tanh(i)
        tanh_value.append(tanh)
                      
    return tanh_value

tanh_line = tanh_activation(random_number)
print(tanh_line)
```

```python
#Visualizing the tanh Activation Function
plt.title("tanh Activation Function")
plt.plot(random_number,tanh_line)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```
